# Data Missing

```python
import pandas as pd
```

```python
import numpy as np
```

```python
d={'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]}
```

```python
df=pd.DataFrame(d)
df
```

```python
df.dropna()# drop the row even a single nan vaule is there, by default axis=0
```

```python
df.dropna(axis=1) # for column
```

```python
df.dropna(thresh=2) #here thresh denote no.of values in that row or col
```

```python
df.fillna(value='FILL')
```

```python
df['A'].fillna(value=df['A'].mean())# fill by mean of col A for col A
```
