```python
import pandas as pd
import numpy as np
```

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.groupby('Company')
```

```python
by_company = df.groupby('Company')
by_company
```

```python
by_company.mean(numeric_only=True) #its only consider numerical columns.
```

```python
by_company.sum()
```

```python
by_company.sum(numeric_only=True)
```

```python
by_company.std(numeric_only=True).loc['FB']
```

```python
df.groupby('Company').std(numeric_only=True).loc['FB']# in single line
```
