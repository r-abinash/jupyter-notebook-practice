# condition

A very important feature of pandas is the ability to perform condition and selection using bracket notation.

```python
import numpy as np
import pandas as pd
```

```python
from numpy.random import randn
```

```python
np.random.seed(101)
```

```python
df=pd.DataFrame(randn(5,4),['A','B','C','D','E'],['W','X','Y','Z'])
```

```python
df
```

```python
booldf = df > 0 # return in true or false basis
```

```python
df[booldf] # when we pass booldf ,true -> values, false -> NaN
```

# By column

```python
df['W']>0 # series
```

```python
df[df['W']>0]
```

```python
# grab all the rows in my data frame where z is less than zero.
df[df['Z']<0]
```

```python
# after filter in the returned subset .I want a particular row or col
result = df[df['W']>0]
```

```python
result
```

```python
result['X']
```

```python
df[df['W']>0]['X']
```

```python
df[df['W']>0][['X','Y','Z']] # pass list to get multiple columns
```

```python
boolseries = df['W']>0
result = df[boolseries]
my_col = ['X','Y','Z']
result[my_col]              # both are same ,but use above one to reduce memory use.
```

# multiple conditions

```python
df[(df['W']>0) and (df['Y']>1)] #(series of boolean<-operator->series of boolean) got confused to over these case use (and -> &, or -> |)
```

```python
df[(df['W']>0) & (df['Y']>1)]
```

```python
df[(df['W']>0) | (df['Y']>1)]
```

# Multi-Index and Index Hierarchy


```python
import pandas as pd
import numpy as np
from numpy.random import randn
```

```python
# Index Levels
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
outside
```

```python
inside
```

```python
list(zip(outside,inside))
```

```python
pd.MultiIndex.from_tuples(hier_index)
```

```python
df = pd.DataFrame(randn(6,2),hier_index,['A','B'])
df # example for multi-index
```

```python
# access data from multi level index, frst acces outside index then go deeper.
df.loc['G1']
```

```python
df.loc['G1'].loc[1] # return in series
```

```python
# give naming for indexies
df.index.names = ['Groups','Num']
df
```

```python
# access single data
df.loc['G2'].loc[1]['A']
```

```python
#cross section it has the ability to skip or go inside a multilevel index
df.xs
```

```python
df.xs('G1')
```

```python
# if i want row 1 fronm both g1 and g2 
df.xs(1,level='Num')
```
