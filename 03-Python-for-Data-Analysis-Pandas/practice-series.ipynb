```python
d = {'a': 1, 'b': 2, 'c': 3}
ser = pd.Series(data=d, index=['a', 'b', 'c']) #The keys of the dictionary match with the Index values, 
ser                                            #hence the Index values have no effect.
```

```python
d = {'a': 1, 'b': 2, 'c': 3}                       #Note that the Index is first build with the keys from the dictionary.
ser = pd.Series(data=d, index=['x', 'y', 'z'])     #After this the Series is reindexed with the given Index values,
ser                                                #hence we get all NaN as a result.
```

```python
r = [1, 2]
ser = pd.Series(r, copy=False)  #copy=true -> make a copy of orginal data
ser.iloc[0] = 999               #copy=false -> not make a copy , just work on the orginal data
r                               # list has no view for Series but numpy array has.
```

```python
ser     #Due to input data type as (list) the Series has a `copy` of
        #the original data even though `copy=False`, so
        #the orginal data is unchanged.
```

```python
arr = np.array(r)
ser = pd.Series(arr,copy=False)
ser.iloc[0] = 999
arr
```

```python
ser     #Due to input data type the Series has a `view` on
        #the original data, so
        #the data is changed as well.
```
