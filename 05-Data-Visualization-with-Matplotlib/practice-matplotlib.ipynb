```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x*2.1232/78
```

```python
x
```

```python
y
```

```python
# funtional
plt.plot(x,y)
plt.xlabel('x_lable')
plt.ylabel('y_lable')
plt.title('frst_visual')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.title('frist')
plt.show()

plt.subplot(1,2,2)
plt.plot(y,x)
plt.title('sec')
plt.show()
```

```python
# object oriented->step:1
fig=plt.figure() # here just create figure object. it's like a imaginary blank canvas.
# step:2
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # here we add axes to canvas, for that we just pass a list.[left,bottom,with,height]-> rnage from 0 to 1.
# step:3 plot on the set of axes.
axes.plot(x,y)
# step:4 add lable on it.
axes.set_xlabel('X_lable')
axes.set_ylabel('Y_lable')
axes.set_title('object oriented')
plt.show()
```

```python
#puting two set of figure in one canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.65,0.3,0.2])

axes1.plot(x,y)
axes2.plot(y,x,'r*')

axes1.set_title('outer plot')
axes2.set_title('inner plot')
plt.show()
```

#### let see how to create subplot by using object oriented


```python
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()# two fix overlapping of plots.
plt.show()

# we have plt.subplot and essentially it's just an axes manager on top of plt.figure
```

```python
axes # here the 'axes' object is list of matplotlib. which are the one we manually created in fig.add_axes(). 
```

```python
# we can travel throght axes array to plots each by 'for' loop or indexing.
# for row in axes:
#     for ax in row:
#         ax.plot(x, y)
#         ax.set_title('sub-plot')

axes[0][1].plot(x,y)
axes[0][1].set_title('check')
fig
```

#### fig size , DPI

```python
fig = plt.figure(figsize=(3,2))#frst argument is figsize pass in tuple(weight,height). second arg is DPI which is dot per inch or pixels per inch.
                               #for now go with default DPI.
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
fig
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[0].set_title('1 st')
axes[1].plot(y,x)
axes[1].set_title('2 nd')
plt.tight_layout()
fig
```

#### Legends

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

# here we have 2 plots. probably a little confused as to what each plot represents.
axes.plot(x,x**2)
axes.plot(x,x**3)

axes.set_xlabel('x_label')
axes.set_ylabel('y_label')
axes.set_title('visual')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

#To slove above legends cames
# two steps for declare legends ,1-> declare label for each axes, 2->then call legend from axes.
axes.plot(x,x**2,label='x square')
axes.plot(x,x**3,label='x cube')

axes.legend(loc=0) # you can also specify the location of legends in axes,best location code is 0 ,and 0 to 10 loc are available 
                   # or u can specify in tuple(left,bottom)-> axes.legend(loc=(0.3,0.3))

axes.set_xlabel('x_label')
axes.set_ylabel('y_label')
axes.set_title('visual')
plt.show()
```

#### plot appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='lavender')# here we can use color name or "RGB hex code" like '#FF8C00'
plt.show()
```

```python
#line width and line style
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',linewidth=1,linestyle='-',marker='o',markersize=5,
        markerfacecolor='black',markeredgewidth=2,markeredgecolor='lavender',alpha=0.9) 
#linewidth(lw) is size of the line, and alpha controls color transparent.
#linestyle(ls) controls how the line wnats to look like(--,-.,: etc..).
# marker -> helps in ploting the points of (x,y) in axes(*,1,+,o)
# markersize is also available 
fig
```

#### control over axis appearance
## plot range

```python
# we can also give range for x and y axes.
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,ls=':')
ax.set_xlim([0,1])#[upper bound, lower bound]
ax.set_ylim([0,2])
fig
```
